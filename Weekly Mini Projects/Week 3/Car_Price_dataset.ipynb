1.1 Project Overview

The project focuses on understanding the factors that influence car prices and building predictive models to estimate vehicle prices accurately. The automotive market is complex, with pricing influenced by multiple features such as brand, model, manufacturing year, mileage, fuel type, transmission, and engine size. By analyzing historical data, we can identify trends, relationships, and key drivers of price, which can assist buyers, sellers, and dealerships in making informed decisions. The project combines data analysis, feature engineering, visualization, and machine learning modeling to achieve these goals.

1.2 Objectives

Perform exploratory data analysis (EDA) to understand data distribution and feature relationships.

Clean and preprocess the dataset to handle missing values, duplicates, and inconsistencies.

Build regression-based machine learning models to predict car prices accurately.

Evaluate and compare model performance using metrics like R², MAE, and RMSE.

Generate actionable insights to help stakeholders in pricing, buying, and selling decisions.

1.3 Problem Statement

Car prices are influenced by a variety of factors that interact in complex ways, including brand reputation, age, mileage, engine type, and overall condition. Predicting prices is challenging due to these multiple variables and their nonlinear relationships. The goal of this project is to develop a predictive model that can accurately estimate car prices based on historical data and relevant features. A successful model can help stakeholders make informed decisions, optimize pricing strategies, and better understand market dynamics.

```python
# Select numerical columns
numerical_cols = ['year', 'price', 'mileage', 'tax', 'mpg', 'engineSize']
```

```python
# Visualize Outliers using Boxplots

for col in numerical_cols:
    plt.figure(figsize=(10, 6))
    sns.boxplot(data[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
    
```

```python
# Removing extreme outliers

for col in numerical_cols:
    q1 = data[col].quantile(0.25)
    q3 = data[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    data = data[(data[col] >= lower_bound) & (data[col] <= upper_bound)]
# Check the shape after removing outliers
print(f"Dataset shape after outlier removal: {data.shape}")
```
