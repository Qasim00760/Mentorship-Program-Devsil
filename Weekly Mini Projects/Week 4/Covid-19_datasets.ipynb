```python
# List of numeric columns
num_cols = ['Total_Cases', 'New_Cases', 'Total_Deaths', 'New_Deaths',
            'People_Vaccinated', 'People_Fully_Vaccinated', 'Total_Boosters']

# Keep only columns that exist in the dataset
num_cols_existing = [col for col in num_cols if col in data.columns]

# Convert to numeric
for col in num_cols_existing:
    data[col] = pd.to_numeric(data[col], errors='coerce')

# Convert categorical columns
cat_cols = ['Country_Code', 'Continent', 'Country']
cat_cols_existing = [col for col in cat_cols if col in data.columns]

for col in cat_cols_existing:
    data[col] = data[col].astype('category')
print(data.dtypes)
```

```python
# Cumulative and daily columns
cum_cols = ['Total_Cases', 'Total_Deaths', 'People_Vaccinated', 'People_Fully_Vaccinated', 'Total_Boosters']
daily_cols = ['New_Cases', 'New_Deaths']

# Keep only existing columns
cum_cols = [c for c in cum_cols if c in data.columns]
daily_cols = [c for c in daily_cols if c in data.columns]

# Apply filling
data[cum_cols] = data[cum_cols].fillna(method='ffill')
data[daily_cols] = data[daily_cols].fillna(0)

print("Missing values after handling:")
print(data.isnull().sum())

# remove missing values
data = data.dropna()
print("Missing values after dropping rows with any missing data:")
```

Rows with missing country identifiers were removed to maintain data integrity.

Missing values in cumulative variables were forward-filled to preserve time-series

continuity while daily metrics were filled with zero where appropriate.

```python
# Columns where negative values are unreasonable
non_negative_cols = [
    'New_Cases',
    'New_Deaths',
    'Total_Cases',
    'Total_Deaths',
    'Total_Boosters'   # may or may not exist
]

# Keep only columns that exist in the dataset
existing_cols = [col for col in non_negative_cols if col in data.columns]

# Apply clipping
for col in existing_cols:
    data[col] = data[col].clip(lower=0)

print("Negative values handled for columns:", existing_cols)
```

Negative values in case and death counts are not logically valid and may occur
due to reporting corrections. Such values were clipped to zero to ensure
data consistency while preserving realistic trends.

```python
# Extract year, month, and week from date
data['Year'] = data['Observation_Date'].dt.year
data['Month'] = data['Observation_Date'].dt.month
data['Week'] = data['Observation_Date'].dt.isocalendar().week
print(data.head())
```

These features allow trend analysis seasonal pattern detection and monthly/weekly comparisons of COVID-19 spread.

```python
# Calculate death rate
if 'Total_Cases' in data.columns and 'Total_Deaths' in data.columns:
    data['Death_Rate'] = data['Total_Deaths'] / data['Total_Cases']
    data['Death_Rate'] = data['Death_Rate'].replace([np.inf, -np.inf], 0).fillna(0)
    print("Death rate calculated.")
else:
    print("Total_Cases or Total_Deaths column missing; cannot calculate Death_Rate.")
    data['Death_Rate'] = data['Death_Rate'].replace([np.inf, -np.inf], 0).fillna(0) 
# Preview final dataset
print(data.head())
```

Death rate measures disease severity and healthcare effectiveness across countries and time.

```python
# Calculate case fatality rate comparison
data['Case_Fatality_Rate'] = data['Death_Rate'] * 100
data
```

Expresses deaths as a percentage of confirmed cases making comparisons easier across regions.

```python
print(data.columns)
```

```python
# Daily growth rate of total cases
if 'Total_Cases' in data.columns:
    data['Case_Growth_Rate'] = (
        data.groupby('Country')['Total_Cases']
        .pct_change()
        .replace([np.inf, -np.inf], 0)
        .fillna(0)
    )
# Preview final dataset
print(data.head())
```

Growth rate helps identify countries experiencing rapid spread.

```python
# Vaccination coverage
if 'People_Vaccinated' in data.columns and 'Total_Cases' in data.columns:
    data['Vaccination_to_Case_Ratio'] = (
        data['People_Vaccinated'] / data['Total_Cases']
    ).replace([np.inf, -np.inf], 0).fillna(0)
# Preview final dataset
print(data.head())
```

This feature helps evaluate the relationship between vaccination campaigns and infection levels.

```python
# Log transformation for highly skewed variables
for col in ['Total_Cases', 'Total_Deaths', 'New_Cases', 'New_Deaths']:
    if col in data.columns:
        data[f'Log_{col}'] = np.log1p(data[col])

# Preview final dataset
print(data.head())
```

Log transformation stabilizes variance and improves visualization and statistical modeling.

```python
# Categorize pandemic intensity
if 'New_Cases' in data.columns:
    data['Pandemic_Phase'] = pd.cut(
        data['New_Cases'],
        bins=[-1, 0, 100, 1000, np.inf],
        labels=['No Cases', 'Low', 'Moderate', 'High']
    )
# Preview final dataset
print(data.head())
```

Converts continuous case numbers into interpretable pandemic severity categories.

```python
# Check if column exists
if 'New_Cases_7Day_Avg' not in data.columns:
    # Drop rows with missing New_Cases
    data = data.dropna(subset=['New_Cases'])

    # Sort by Country and Date
    data = data.sort_values(['Country', 'Observation_Date'])

    # Create 7-day rolling average
    data['New_Cases_7Day_Avg'] = (
        data.groupby('Country')['New_Cases']
        .rolling(window=7, min_periods=1)
        .mean()
        .reset_index(level=0, drop=True)
    )
```

```python
print(data[['Year','Month','Death_Rate','Case_Fatality_Rate',
            'New_Cases_7Day_Avg','Case_Growth_Rate']].head())
```

Feature engineering enhanced the dataset by extracting temporal attributes
calculating death and growth rates smoothing trends using rolling averages
and transforming skewed variables. These engineered features improve trend
analysis cross country comparisons and the evaluation of vaccination impact.

```python
from scipy.stats import pearsonr
import seaborn as sns
import matplotlib.pyplot as plt

# Drop missing values first
sample = data[['New_Cases', 'New_Deaths']].dropna()

# Check if enough data exists
if len(sample) > 10:
    sample = sample.sample(min(1000, len(sample)), random_state=42)

    corr, p = pearsonr(sample['New_Cases'], sample['New_Deaths'])

    sns.scatterplot(x='New_Cases', y='New_Deaths', data=sample)
    plt.title(f"Cases vs Deaths (Correlation = {corr:.2f})")
    plt.show()

    print("Correlation:", corr)
    print("P-value:", p)
else:
    print("Not enough data for correlation analysis.")
```

Statistically significant relationship

Increases in COVID-19 cases are strongly associated with increases in deaths.

# Pre vs Post-Vaccination Death Comparison

```python
median_date = data['Observation_Date'].median()

before = data[data['Observation_Date'] < median_date]['New_Deaths']
after = data[data['Observation_Date'] >= median_date]['New_Deaths']

from scipy.stats import ttest_ind
t_stat, p_val = ttest_ind(before.dropna(), after.dropna())

print("P-value:", p_val)
```

**Interpretation**

Lower deaths after vaccination period

Suggests vaccination effectiveness
