# Project Title: Cricket Score analysis

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
import pandas as pd
from io import StringIO

data = """Match,Team,Runs,Wickets,Overs
1,Pakistan,250,8,50
2,India,275,6,50
3,Australia,300,7,50
4,England,280,9,50
5,Pakistan,260,10,49.3
6,India,290,5,50
7,Australia,310,6,50
8,England,270,8,50"""

df = pd.read_csv(StringIO(data))
print(df)
```

```python
# Inspect data
print("First 5 Rows:")
df.head()
print("\nDataset Info:")
df.info()
```

```python
# Basic statistics
print("\nStatistical Summary:")
df.describe()
```

```python
# Mean, Min, Max Runs
mean_runs = df['Runs'].mean()
min_runs = df['Runs'].min()
max_runs = df['Runs'].max()

print("Mean Runs:", mean_runs)
print("Minimum Runs:", min_runs)
print("Maximum Runs:", max_runs)
```

```python
# Group by team
team_performance = df.groupby("Team")["Runs"].mean()
print("\nAverage Runs by Team:")
print(team_performance)
```

```python
import matplotlib.pyplot as plt

# Group by team (from previous cell)
team_performance = df.groupby("Team")["Runs"].mean()

# Visualization
plt.figure(figsize=(8, 6))
team_performance.plot(kind='bar')
plt.title("Average Runs by Team")
plt.xlabel("Team")
plt.ylabel("Average Runs")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
